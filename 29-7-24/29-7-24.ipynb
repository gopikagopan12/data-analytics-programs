```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
p_win = 0.2
p_lose = 0.8
n = 10
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p_win)
plt.bar(x, pmf, color='blue')
plt.xlabel('Number of Wins')
plt.ylabel('Probability')
plt.title('PMF of Number of Wins in 10 Games')
plt.show()

```

### Probability of Hitting the Target at Least Twice


```python
from scipy.stats import binom
p_hit = 1/4
n_shots = 7
prob_at_least_two_hits = 1 - binom.cdf(1, n_shots, p_hit)

print(f"Probability of hitting the target at least twice: {prob_at_least_two_hits:.4f}")
```

###  Poisson Distribution Example and Histogram

```python
from scipy.stats import poisson
lambda_poisson = 3
x = np.arange(0, 15)
pmf_poisson = poisson.pmf(x, lambda_poisson)
plt.bar(x, pmf_poisson, color='green')
plt.xlabel('Number of Occurrences')
plt.ylabel('Probability')
plt.title('PMF of Poisson Distribution (λ=3)')
plt.show()





```

### Normal Distribution for Heights

```python
from scipy.stats import norm
mean_height=160
std_height=40
x = np.linspace(mean_height - 4*std_height, mean_height + 4*std_height, 1000)
pdf=norm.pdf(x,mean_height,std_height)
plt.plot(x,pdf,color='blue')
plt.xlabel('height(cm)')
plt.ylabel('density')
plt.title('distribution of heights')
plt.show()
```

### central limit theorem

```python
num=[1,10,50,100]
means=[]
for j in num:
    np.random.seed(1)
    x=[np.mean(np.random.randint(-40,40,j))for i in range(1000)]
    means.append(x)
k=0
fig,ax=plt.subplots(2,2,figsize=(8,8))
for i in range(0,2):
    for j in range(0,2):
        ax[i,j].hist(means[k],10,density=True)
        ax[i,j].set_title(label=num[k])
        k=k+1
plt.show()
```

### Sampling Methods on Iris Dataset

```python
pip install scikit-learn
```

```python
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris
```

```python
import pandas as pd
```

```python
iris_df=pd.DataFrame(iris.data)
```

```python
iris_df.head()
```

```python
iris_df['class']=iris.target
```

```python
iris_df.head()
```

```python
iris_df.columns=['sepal_len','sepal_wid','petal_len','petal_wid','class']
```

```python
iris_df.head()
```

```python
iris_df['class'].value_counts()
```

```python
X=iris_df.drop(columns=['class'])
y=iris_df['class']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(X,y,test_size=0.25,random_state=None,shuffle=True,stratify=None)
```

```python
print("Training classes: ",ytrain.value_counts())
print("Testing classes", ytest.value_counts())
```

```python
ytest.value_counts()
```

```python
xtrain,xtest,ytrain,ytest=tts(X,y,test_size=0.25,random_state=10,shuffle=True,stratify=None)
print("Training classes: ",ytrain.value_counts())
print("Testing classes", ytest.value_counts())
```

### Stratified Sampling

```python
xtrain,xtest,ytrain,ytest=tts(X,y,test_size=0.2,random_state=10,shuffle=True,stratify=None)
print("Training classes: ",ytrain.value_counts())
print("Testing classes", ytest.value_counts())
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
skf=StratifiedShuffleSplit(n_splits=3,train_size=0.8)
```

```python
for i,(train_index,test_index) in enumerate(skf.split(X,y)):
    print("No. of folds",i)
    print(iris_df.iloc[train_index]['class'].value_counts())
    print(iris_df.iloc[test_index]['class'].value_counts())
```
